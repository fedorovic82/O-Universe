```
from pyspark.sql import functions as f
```

```
# Location where to put the table in the Databricks database menu
hiveTable = "dev_derived_ouniverse.output_total_universe_nl_enriched"
```

```
dfU = spark.table(hiveTable)
```

```
display(dfU)
```

```
dfM = spark.table("dev_sources_scraper.intermediate_menu_data") \
  .where(f.lower(f.col('country')).contains('netherlands')) \
  .where(f.lower(f.col('text')).contains("terras")) \
  .withColumn("menuHasTerras", f.lit(True)) \
  .select('id', 'menuHasTerras') \
  .distinct()
```

```
display(dfM)
```

```
dfM.select(f.col('id').substr(1,4).alias('idSub')).groupBy('idSub').count().show()
```

```
dfE = dfU.join(dfM, ((f.trim(f.regexp_replace(dfM.id, '[/\s]+', '')) == f.trim(f.regexp_replace(f.col('operatorId'), '[/\s]+', ''))) | (dfM.id==dfU.PlaceIDGoogle)), how='left').drop(dfM.id).drop(dfU.menuHasTerras)
```

```
dfM.where(f.col('menuHasTerras').isNotNull()).count()
```

```
dfE.where(f.col('menuHasTerras').isNotNull()).count()
```

```
# Location where to save the Delta Table in the DBFS
deltaTable = "/mnt/datamodel/dev/derived/ouniverse/output/total_universe_nl_enriched"

# Write the data to a Delta Table
dfE.write \
  .format("delta") \
  .mode("overwrite") \
  .option("overwriteSchema", "true") \
  .save(deltaTable)

# Load the Delta Table into the Data interface within Databricks.
sqlQuery1 = "drop table if exists " + hiveTable
sqlQuery2 = "create table " + hiveTable + " using delta location " + "'" + deltaTable + "'"
spark.sql(sqlQuery1)
spark.sql(sqlQuery2)
```
