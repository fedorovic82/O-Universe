```
# https://download.geofabrik.de/
import requests

class DownloadOsmData(object): 
  def __init__(self, continent, country, storageFolder = '/dbfs/mnt/datamodel/dev/sources/osm/raw/'):
    self.continent = continent
    self.country = country
    self.storageFolder = storageFolder

  def download_osm_data(self):
    url = 'https://download.geofabrik.de/' + self.continent + '/' + self.country + '-latest.osm.pbf'
    r = requests.get(url)
    fileNameLocal = self.storageFolder + self.country + '.osm.pbf'
    print("Writing File")
    with open(fileNameLocal, 'wb') as f:
      f.write(r.content)
    print("Writing Done")
```

```
continent = getArgument("continent")
country = getArgument("country")
DownloadOsmData(continent, country).download_osm_data()
```
